### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 3. Calificación promedio de la tienda


```python
# Calcular promedio de calificación en cada tienda
calificaciones = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}
```

```python
# Mostrar los resultados con formato
print("=== Calificación promedio por tienda ===")
for nombre, promedio in calificaciones.items():
    print(f"{nombre}: {promedio:.2f}")
```

```python
#Visualización: comparación de calificaciones

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(calificaciones.keys(), calificaciones.values(), color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio (1 a 5)")
plt.ylim(0, 5)  # rango fijo de 1 a 5 para facilitar comparación
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```

#####Interpretación

Todas las tiendas se encuentran en un rango bastante homogéneo (3.9–4.1), lo cual indica una satisfacción constante y buena por parte de los clientes.

La Tienda 3 es la mejor valorada, aunque por muy poco margen.

La Tienda 1 podría revisar detalles de servicio, calidad o logística para subir ese promedio a más de 4.1.

En la práctica empresarial, cuando todas las tiendas tienen valores cercanos a 4.0, eso muestra consistencia operacional, lo que es una fortaleza importante ✅


# 4. Productos más y menos vendidos

```python
#value_counts()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Lista de tiendas y nombres (asegúrate de que estas variables existan) ---
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# DataFrame para resumen final
resumen = []

for df, nombre in zip(tiendas, nombres):
    print(f"\n=== {nombre} ===")

    # 1) Contar productos (elimina NaN por seguridad)
    conteo = df['Producto'].dropna().value_counts()

    # 2) Top y Bottom (manejo de empates)
    max_count = conteo.max()
    min_count = conteo.min()

    # Productos más vendidos (puede haber varios con la misma cantidad máxima)
    mas_vendidos = conteo[conteo == max_count]
    # Productos menos vendidos (puede haber varios con la misma cantidad mínima)
    menos_vendidos = conteo[conteo == min_count]

    print("Productos más vendidos (nombre: ventas):")
    print(mas_vendidos.to_string(), "\n")

    print("Productos menos vendidos (nombre: ventas):")
    print(menos_vendidos.to_string(), "\n")

    # Añadir al resumen (tomamos los top 1 y bottom 1 para resumen; si hay empates, se toman todos en listas)
    resumen.append({
        "tienda": nombre,
        "top_products": list(mas_vendidos.index),
        "top_count": int(max_count),
        "bottom_products": list(menos_vendidos.index),
        "bottom_count": int(min_count)
    })

    # 3) Visualización: Top 10 y Bottom 10
    # Top 10
    plt.figure(figsize=(10,4))
    conteo.head(10).plot(kind='bar', title=f"Top 10 productos - {nombre}")
    plt.ylabel("Cantidad vendida")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Bottom 10 (invertir para que el más pequeño aparezca arriba)
    # Si hay menos de 10 productos, muestra todos
    bottom_n = conteo.tail(10)
    plt.figure(figsize=(10,4))
    bottom_n[::-1].plot(kind='barh', title=f"Bottom 10 productos - {nombre}")  # horizontal para mejor lectura
    plt.xlabel("Cantidad vendida")
    plt.tight_layout()
    plt.show()

# Convertir resumen a DataFrame para verlo de forma tabular
resumen_df = pd.DataFrame(resumen)
display(resumen_df)
```

# 5. Envío promedio por tienda

```python
# Calcular promedio de costo de envío por tienda

envio_promedio = {
    "Tienda 1": tienda["Costo de envío"].mean().round(2),
    "Tienda 2": tienda2["Costo de envío"].mean().round(2),
    "Tienda 3": tienda3["Costo de envío"].mean().round(2),
    "Tienda 4": tienda4["Costo de envío"].mean().round(2)
}

print("=== Costo de Envío Promedio por Tienda ===")
for nombre, promedio in envio_promedio.items():
    print(f"{nombre}: ${promedio:,.2f}")
```

```python
import matplotlib.pyplot as plt

nombres = list(envio_promedio.keys())
valores = list(envio_promedio.values())

plt.barh(nombres, valores, color="teal")
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Costo promedio ($COP)")
plt.ylabel("Tienda")
plt.gca().invert_yaxis()  # Tienda 1 arriba
plt.show()
```
